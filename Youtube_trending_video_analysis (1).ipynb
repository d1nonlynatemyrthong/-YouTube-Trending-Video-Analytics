```python
import pandas as pd
import kagglehub
path=kagglehub.dataset_download("datasnaek/youtube-new")
print("Path to datsets:", path)
```

```python
import os
files=os.listdir(path)
print(files)
```

```python
a=os.path.join(path,"GBvideos.csv")
gb=pd.read_csv(a)
b=os.path.join(path,"INvideos.csv")
inv=pd.read_csv(b)
c=os.path.join(path,"USvideos.csv")
us=pd.read_csv(c)
```

```python
print(gb.info())
gb.tail()
```

```python
us.info()
us.tail()
```

```python
inv.info()
inv.tail()
```

```python
us["Country"]="US"
inv["Country"]="IN"
gb["Country"]="Gb"

df=pd.concat([us,inv,gb],ignore_index=True)
df.info()
```

```python
df.head()
```

```python
df["trending_date"]=pd.to_datetime(df["trending_date"],format='%y.%d.%m',errors='coerce')
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
```

```python
df["publish_date"]=df["publish_time"].dt.date
df["publish_hour"]=df["publish_time"].dt.hour
df["video_unique"]=df["video_id"]+"_"+df["Country"]
```

```python
trend_counts = df.groupby('video_unique')['trending_date'].nunique().reset_index()
trend_counts.columns = ['video_unique', 'trending_days']
df = df.merge(trend_counts, on='video_unique', how='left')
```

```python
df["publish_to_trend"]=(df["trending_date"]-pd.to_datetime(df["publish_date"])).dt.days
```

```python
df['title'] = df['title'].fillna('')
df['tags'] = df['tags'].fillna('')
df['description']=df['description'].fillna('')
```

```python
df.info()
```

```python
from textblob import TextBlob
df["polarity"]=df["title"].apply(lambda x:TextBlob(x).sentiment.polarity)
df["tags_polarity"]=df["tags"].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
def get_sentiment(score):
    if score>0.2:
        return "Positive"
    elif score<-0.2:
        return 'Negative'
    else:
        return "Neutral"
df['title_sentiment'] = df['polarity'].apply(get_sentiment)
df['tags_sentiment'] = df['tags_polarity'].apply(get_sentiment)
```
