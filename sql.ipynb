```python
import sqlite3
```

```python
import pandas as pd
df = pd.read_csv("youtube_trending_cleaned.csv")
conn=sqlite3.connect("youtube_trending.db")
df.to_sql("youtube_data",conn,index=False)
```

```python
query1 = """
SELECT country, category_id, ROUND(AVG(views), 0) AS avg_views
FROM youtube_data
GROUP BY country, category_id
ORDER BY country, avg_views DESC;
"""
views_by_category = pd.read_sql_query(query1, conn)
views_by_category.to_csv("views_by_category_country.csv", index=False)
views_by_category.head()
```

```python
query2 = """
SELECT country, category_id,
       ROUND(AVG(likes), 0) AS avg_likes,
       ROUND(AVG(dislikes), 0) AS avg_dislikes
FROM youtube_data
GROUP BY country, category_id
ORDER BY country, avg_likes DESC;
"""
likes_dislikes = pd.read_sql_query(query2, conn)
likes_dislikes.to_csv("likes_dislikes_by_category.csv", index=False)
likes_dislikes.head()
```

```python
query3 = """
SELECT country, title_sentiment, COUNT(*) AS total_videos
FROM youtube_data
GROUP BY country, title_sentiment
ORDER BY country, total_videos DESC;
"""
sentiment_count = pd.read_sql_query(query3, conn)
sentiment_count.to_csv("sentiment_counts.csv", index=False)
sentiment_count.head()
```

```python
query4 = """
SELECT country, ROUND(AVG(publish_to_trend), 1) AS avg_days_to_trend
FROM youtube_data
WHERE publish_to_trend >= 0
GROUP BY country;
"""
days_to_trend = pd.read_sql_query(query4, conn)
days_to_trend.to_csv("avg_days_to_trend.csv", index=False)
days_to_trend.head()
```

```python
query5 = """
SELECT country, category_id, ROUND(AVG(trending_days), 1) AS avg_trending_days
FROM youtube_data
GROUP BY country, category_id
ORDER BY country, avg_trending_days DESC;
"""
trending_duration = pd.read_sql_query(query5, conn)
trending_duration.to_csv("trending_duration_by_category.csv", index=False)
trending_duration.head()
```

```python
conn.close()
```
